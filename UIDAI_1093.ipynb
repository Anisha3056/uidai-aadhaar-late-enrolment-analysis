# **Notebook Section: State Name Standardization & Validation**

Standardize inconsistent state names, correct legacy and spelling variants, remove invalid locality-level values, and ensure that the state field contains only valid Indian states and union territories.

**Display the list of state names in the dataset**

```python
merged_clean['state'].unique()
```

**Remove Clearly Invalid Locality-Level Entries**

```python
# Locality or city names incorrectly present in the state column
invalid_states = ["BALANAGAR","Darbhanga","Jaipur","Madanapalle","Puttenahalli","Nagpur","Raja Annamalai Puram"]

merged_clean = merged_clean[~merged_clean["state"].isin(invalid_states)].copy()
merged_clean.shape
```

11 records were removed with locality or city names mentioned as state names.

**Define Canonical List of Valid States**

```python
# Official list of Indian states and union territories
valid_states = [
    "Andhra Pradesh","Arunachal Pradesh","Assam","Bihar","Chhattisgarh","Goa","Gujarat",
    "Haryana","Himachal Pradesh","Jharkhand","Karnataka","Kerala","Madhya Pradesh",
    "Maharashtra","Manipur","Meghalaya","Mizoram","Nagaland","Odisha","Punjab","Rajasthan",
    "Sikkim","Tamil Nadu","Telangana","Tripura","Uttar Pradesh","Uttarakhand","West Bengal",
    "Andaman and Nicobar Islands","Chandigarh","Dadra and Nagar Haveli and Daman and Diu",
    "Delhi","Jammu and Kashmir","Ladakh","Lakshadweep","Puducherry"
]
```

**Normalize State Names**

```python
import re

def normalize_state(s):
    s = s.lower().strip()
    s = re.sub(r"[^a-z\s]", "", s)   # Remove punctuation and symbols
    s = re.sub(r"\s+", " ", s)       # Normalize whitespace
    return s

merged_clean["state_norm"] = merged_clean["state"].astype(str).apply(normalize_state)
```

**Map Known Legacy and Variant Names**

```python
# Manual mapping for known legacy names and common variants
alias_map = {
    "west bengal": "West Bengal",
    "westbengal": "West Bengal",
    "west bangal": "West Bengal",
    "west bengli": "West Bengal",

    "andhra pradesh": "Andhra Pradesh",
    "tamilnadu": "Tamil Nadu",

    "uttaranchal": "Uttarakhand",
    "orissa": "Odisha",
    "pondicherry": "Puducherry",

    "daman and diu": "Dadra and Nagar Haveli and Daman and Diu",
    "dadra and nagar haveli": "Dadra and Nagar Haveli and Daman and Diu"
}

merged_clean["state_mapped"] = merged_clean["state_norm"].replace(alias_map)
```

**Fuzzy Match Remaining Variants**

```python
!pip install rapidfuzz # install necessary library for fuzzy matching
```

```python
from rapidfuzz import process, fuzz

valid_states_norm = [normalize_state(s) for s in valid_states]
norm_to_clean = {normalize_state(s): s for s in valid_states}

def fuzzy_map(s):
    match, score, _ = process.extractOne(s, valid_states_norm, scorer=fuzz.ratio)
    if score >= 85:
        return norm_to_clean[match]
    return None

mask = ~merged_clean["state_mapped"].isin(valid_states)
merged_clean.loc[mask, "state_mapped"] = merged_clean.loc[mask, "state_mapped"].apply(fuzzy_map)
```

**Remove Unmatched Records and Finalize**

```python
# Remove records that could not be mapped to any valid state
merged_clean = merged_clean[merged_clean["state_mapped"].notna()]

# Replace original state column with cleaned version
merged_clean["state"] = merged_clean["state_mapped"]

# Drop helper columns
merged_clean = merged_clean.drop(columns=["state_norm", "state_mapped"])
```

**Validate**

```python
merged_clean["state"].value_counts()
len(merged_clean["state"].unique())
```

The state field exhibited significant inconsistencies, including variations in casing, spelling errors, legacy names (e.g., Orissa, Pondicherry), merged or deprecated administrative units (e.g., Daman and Diu), and invalid locality-level entries (e.g., city or neighborhood names).

To ensure semantic consistency, we standardized state names through a multi-step process: normalization, manual alias mapping for known variants, fuzzy string matching against an official list of Indian states and union territories, and removal of unmappable or invalid entries. This resulted in a validated set of 36 unique administrative units, enabling accurate geographic aggregation and comparison.

This ensures consistency for all state-level aggregations and prevents double-counting due to spelling or formatting differences.

# **Intra-State Enrollment Inequality (Pincode-Level Analysis)**

**Aggregate Enrollment at StatePincode Level**

```python
# Aggregate total Aadhaar enrollment by state and pincode
pincode_state = (
    merged_clean
    .groupby(["state", "pincode"])["total_enrollment"]
    .sum()
    .reset_index()
)
```

**Compute Dispersion of Enrollment Within Each State**

```python
# Compute mean and standard deviation of enrollment across pincodes for each state
pincode_dispersion = (
    pincode_state
    .groupby("state")["total_enrollment"]
    .agg(["mean", "std"])
)
```

**Calculate Coefficient of Variation (CV)**

```python
# Coefficient of Variation measures relative dispersion:
# higher CV = more unequal distribution across pincodes
pincode_dispersion["cv"] = pincode_dispersion["std"] / pincode_dispersion["mean"]

# Sort states by highest inequality
pincode_dispersion = pincode_dispersion.sort_values("cv", ascending=False)

pincode_dispersion.head(10)
```

**Visualize Top 10 States by Intra-State Inequality**

```python
# Select top 10 states with highest dispersion
top10_disp = pincode_dispersion.head(10)

# Plot the coefficient of variation
plt.figure(figsize=(10,5))
plt.bar(top10_disp.index, top10_disp["cv"])
plt.xticks(rotation=80)
plt.ylabel("Coefficient of Variation (Dispersion)")
plt.title("Top 10 States with Highest Intra-State Enrollment Inequality")
plt.tight_layout()
plt.show()
```

To examine geographic inequality in Aadhaar enrollment within states, we aggregated enrollment counts at the pincode level and computed the coefficient of variation (CV) across pincodes for each state. The CV captures relative dispersion  higher values indicate that enrollment is concentrated in a few pincodes, while lower values indicate a more even spatial distribution.

States such as Punjab, Uttarakhand, and Haryana exhibit high intra-state dispersion, suggesting that Aadhaar activity may be concentrated in select urban or administrative centers, potentially leaving peripheral or rural areas underserved. This highlights the need for more geographically balanced outreach and service delivery.

```python
ne_trend = merged_clean[merged_clean["region"]=="North-East"] \
    .groupby(merged_clean["date"].dt.month)[["age_18_greater"]].sum()

assam_trend = merged_clean[merged_clean["state"]=="Assam"] \
    .groupby(merged_clean["date"].dt.month)[["age_18_greater"]].sum()

# Create a dataframe comparing total North-East late enrolments
# with Assam's late enrolments month-wise
monthly_late_enrollment_comparison = pd.DataFrame({
    "North_East_total_late_enrollment": ne_trend["age_18_greater"],
    "Assam_late_enrollment": assam_trend["age_18_greater"]
})

# Compute what percentage of North-East's late enrolments
# come specifically from Assam for each month
monthly_late_enrollment_comparison["Assam_percentage_of_NE_late_enrollment"] = (
    monthly_late_enrollment_comparison["Assam_late_enrollment"] /
    monthly_late_enrollment_comparison["North_East_total_late_enrollment"]
) * 100

# Display the comparison table
display(monthly_late_enrollment_comparison)
```

**Monthly State-wise Contribution to Late Enrolment in North-East**

```python
# Initialize a dataframe with total North-East late enrolments per month
ne_states_late_enrollment_comparison = pd.DataFrame({
    "North_East_total_late_enrollment": ne_trend["age_18_greater"]
})

# For each North-Eastern state, compute its monthly late enrolment
for state in north_east_states:
    state_trend = merged_clean[merged_clean["state"] == state] \
        .groupby(merged_clean["date"].dt.month)[["age_18_greater"]].sum()

    # Add the state's monthly late enrolment as a new column
    ne_states_late_enrollment_comparison[f"{state}_late_enrollment"] = \
        state_trend["age_18_greater"]

# Convert absolute late enrolments into percentage contribution
# of each state relative to total North-East late enrolments
for state in north_east_states:
    ne_states_late_enrollment_comparison[
        f"{state}_percentage_of_NE_late_enrollment"
    ] = (
        ne_states_late_enrollment_comparison[f"{state}_late_enrollment"] /
        ne_states_late_enrollment_comparison["North_East_total_late_enrollment"]
    ) * 100

# Extract only percentage columns for cleaner display
percentage_columns = [
    col for col in ne_states_late_enrollment_comparison.columns
    if "_percentage_of_NE_late_enrollment" in col
]

# Display percentage contribution table (replace NaN with 0 for clarity)
display(ne_states_late_enrollment_comparison[percentage_columns].fillna(0))
```

```python
# Calculate total late enrollment for each North-East state
ne_states_total_late_enrollment = merged_clean[merged_clean['state'].isin(north_east_states)] \
                                     .groupby('state')['age_18_greater'].sum()

# Calculate the total late enrollment for all North-East states combined
total_ne_late_enrollment = ne_states_total_late_enrollment.sum()

# Calculate the percentage contribution of each state
ne_states_percentage = (ne_states_total_late_enrollment / total_ne_late_enrollment) * 100

# Filter out states with 0% contribution for cleaner visualization, if any
ne_states_percentage = ne_states_percentage[ne_states_percentage > 0]

# Plotting the bar chart
plt.figure(figsize=(12, 6)) # Adjust figure size for bar chart
bars = plt.bar(ne_states_percentage.index, ne_states_percentage.values, color='skyblue')
plt.xlabel('State', fontsize=12)
plt.ylabel('Percentage Contribution to NE Late Enrolment', fontsize=12)
plt.title('Percentage Contribution of Each North-East State to Total Late Aadhaar Enrolments', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
